Load the MusicGen model

```python
from audiocraft.models import musicgen
from audiocraft.data.audio import audio_write

# Load the small MusicGen model
model = musicgen.MusicGen.get_pretrained('facebook/musicgen-small')

# Move model to GPU if available
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"

# Check if the model has a `device` attribute or a similar mechanism
# to specify the device and, if so, set it accordingly.
# For example:
if hasattr(model, 'device'):
    model.device = torch.device(device)
elif hasattr(model, 'set_device'):  # Or other potential methods
    model.set_device(torch.device(device))
else:
    print(f"Warning: Could not move model to {device}. Model might run on CPU.")
```

Generate Music from Lyrics/Description

```python
# Input lyrics or description
lyrics = input("Enter a description or lyrics for your song: ")

# Generate music (duration is in seconds)
model.set_generation_params(duration=10)  # Generates 10 seconds of music
wav = model.generate([lyrics])  # You can input multiple prompts

# Save and play
audio_path = "/content/generated_song"
audio_write(audio_path, wav[0].cpu(), model.sample_rate)
```
